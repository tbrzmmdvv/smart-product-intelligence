# Milestone 6 — Diffusion & Generative AI

Generate alternative product/lifestyle images with Stable Diffusion.

```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

from src.data import artifacts_dir, load_splits
from src.utils import setup_notebook_path

setup_notebook_path()
```

```python
splits = load_splits()
product = splits["train_products"].iloc[0]
title = str(product["title"])
description = str(product.get("description", ""))[:200]
prompt = f"Professional e-commerce hero photo, studio lighting, {title}. {description}"
print(prompt)
```

```python
dtype = torch.float16 if torch.cuda.is_available() else torch.float32
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype,
    safety_checker=None,
)
if torch.cuda.is_available():
    pipe = pipe.to("cuda")

image = pipe(prompt, num_inference_steps=25, guidance_scale=7.5).images[0]
out_path = artifacts_dir() / "generated_hero.png"
image.save(out_path)

plt.imshow(image)
plt.axis("off")
plt.title("Generated hero image")
plt.show()
```

## Reflection
- **Works well:** clean product-style prompts with title + short description.
- **Failure modes:** wrong packaging text, unrealistic textures, category drift when description is vague.
- **Value:** marketing placeholder images, not ground-truth product photos.
